```python
#importing the necessary libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
crime=pd.read_csv("crime_data.csv")
crime
```

```python
#normalizatin function min max
def norm_func(i):
    x=(i-i.min())/(i-i.max()-(i-i.min()))
    return (x)
```

```python
crime_norm=norm_func(crime.iloc[:,1:])
crime_norm
```

```python
#creating cluster
crime_c = AgglomerativeClustering(n_clusters=4,affinity='euclidean', linkage='single')
crime_c
```

```python
#save clusters for chart
crime_nc = crime_c.fit_predict(crime_norm)
crime_nc
```

```python
Clusters=pd.DataFrame(crime_nc,columns=['Cluster_hi'])
Clusters
```

```python
crime['clusterid_hi'] = crime_c.labels_
```
